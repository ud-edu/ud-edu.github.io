Stelle die Ergebnisse aus Aufgabe 1 bzw. 2 als Säulendiagramm graphisch dar, indem du das folgende Python-Programm entsprechend abänderst. Ändere ruhig auch die Größe des Diagramms, die Beschriftungen und die Farbe der Säule ab, bis du zufrieden mit dem Ergebnis bist.
